# Data wrangling

## Subsetting the data to columns with the survey questions, i.e. that start with FWB (columns 8 - 17)

```r
financialWB1 <- financialWB[, 8:17]
```

## Checking for NAs

```r
sum(is.na(financialWB1))
```

## Removing rows with NAs if needed

```r
financialWB1 = NaRV.omit(financialWB1)
```

# Testing assumptions

## 1. Sample size

```r
nrow(financialWB1)
```

### *There are 6394 observations (>300), so the size assumption is met.*

## 2. Absence of multicollinearity

```r
# Correlation matrix
financialWBmatrix <- cor(financialWB1)
View(round(financialWBmatrix, 2))
```

### *There are no items with a correlation > .9, so the absence of multicollinearity assumpion is met* 

## 3. There is some relationship between survey items

### *There are no values with a correlation < .3, so this assumption is met as well*

## 4. Bartletts's Test to confirm findings from the correlation matrix

```r
cortest.bartlett(financialWB1)
```

### *The p-value is 0 (<.05), so the test is significant, confirming the findings from the correlation matrix*

## 4. Checking determinants

```r
det(financialWBmatrix)
```

### *This value is > .00001, so there is sufficient relation between the variables to proceed with a factor analysis*

# Factor Analysis

## Run basic principal components analysis with high number of factors and without rotation

```r
pcModel1 <- principal(financialWB1, nfactors = 10, rotate = "none")
pcModel1
```

### *SS Loadings > 1 are PC1 and PC2*

## Examine the Scree Plot

```r
plot(pcModel1$values, type="b")
```

### *This confirms that there are probably 2 factors*

## Run second principal components analysis with suspected number of factors

```r
pcModel2 <- principal(financialWB1, nfactors = 2, rotate = "none")
pcModel2
```

## Examine second model fit through the residuals

```r
residuals <- factor.residuals(financialWBmatrix, pcModel2$loadings)
residuals <- as.matrix(residuals[upper.tri(residuals)])
largeResid <- abs(residuals) > .05
head(largeResid)
```

```r
sum(largeResid)
sum(largeResid/nrow(residuals))
```

### *37% of the residuals are large (<50%), so this is an adequate model fit for the data*

## Rotate factors to determine where each survey item fits

## Oblique rotation

### *Oblique Rotation is used when it is safe to assume that the survey items are conceptually related to each other.* 

```r
pcModel3 <- principal(financialWB1, nfactors = 2, rotate = "oblimin")
pcModel3
```

```r
display_png(file = '/Users/silviacatalina/Google Drive/BethelTech/DS0103-MetricsDataProcessing/L05/codebook.png')
```

### *Observations*
* TC1 seems to be related to not having the ability to manage mney well
* TC2 seems to be related to being able to manage money well 

### Loadings >.3, sorted in descending order - useful when you have a larger number of factors

```r
print.psych(pcModel3, cut = .3, sort=TRUE)
```

## Orthogonal Rotation

### *Assumes that the survey items are not related to each other, so less likely to provide the best fit*

```r
pcModel4 <- principal(financialWB1, nfactors = 2, rotate = "varimax")
pcModel4
```

```r
print.psych(pcModel4, cut=.3, sort=TRUE)
```

```r
display_png(file = '/Users/silviacatalina/Google Drive/BethelTech/DS0103-MetricsDataProcessing/L05/codebook2.png')
```

### *The same factors emerge with orthogonal rotation*

### 2. Reverse Coding Items that are negative

#### *It is important that, when you look at the reliability of a scale, you make sure that all of your items/questions are facing the same the direction, meaning that when you have a certain answer for any of the questions, say a 4 in a 1-5 scale, it means the same thing for all the questions. In this example we have some questions that are related to high levels of finantial well being, and some that are related to low levels of financial wellbeing. So, we need to reverse the scales for one of those factors in order to examine the survey's reliability properly.*

```r
financialWB1$FWB1_3r <- NA
financialWB1$FWB1_3r[financialWB1$FWB1_3 == 1] <- 5
financialWB1$FWB1_3r[financialWB1$FWB1_3 == 2] <- 4
financialWB1$FWB1_3r[financialWB1$FWB1_3 == 3] <- 3
financialWB1$FWB1_3r[financialWB1$FWB1_3 == 4] <- 2
financialWB1$FWB1_3r[financialWB1$FWB1_3 == 5] <- 1

financialWB1$FWB1_5r <- NA
financialWB1$FWB1_5r[financialWB1$FWB1_5 == 1] <- 5
financialWB1$FWB1_5r[financialWB1$FWB1_5 == 2] <- 4
financialWB1$FWB1_5r[financialWB1$FWB1_5 == 3] <- 3
financialWB1$FWB1_5r[financialWB1$FWB1_5 == 4] <- 2
financialWB1$FWB1_5r[financialWB1$FWB1_5 == 5] <- 1

financialWB1$FWB1_6r <- NA
financialWB1$FWB1_6r[financialWB1$FWB1_6 == 1] <- 5
financialWB1$FWB1_6r[financialWB1$FWB1_6 == 2] <- 4
financialWB1$FWB1_6r[financialWB1$FWB1_6 == 3] <- 3
financialWB1$FWB1_6r[financialWB1$FWB1_6 == 4] <- 2
financialWB1$FWB1_6r[financialWB1$FWB1_6 == 5] <- 1

financialWB1$FWB2_1r <- NA
financialWB1$FWB2_1r[financialWB1$FWB2_1 == 1] <- 5
financialWB1$FWB2_1r[financialWB1$FWB2_1 == 2] <- 4
financialWB1$FWB2_1r[financialWB1$FWB2_1 == 3] <- 3
financialWB1$FWB2_1r[financialWB1$FWB2_1 == 4] <- 2
financialWB1$FWB2_1r[financialWB1$FWB2_1 == 5] <- 1

financialWB1$FWB2_3r <- NA
financialWB1$FWB2_3r[financialWB1$FWB2_3 == 1] <- 5
financialWB1$FWB2_3r[financialWB1$FWB2_3 == 2] <- 4
financialWB1$FWB2_3r[financialWB1$FWB2_3 == 3] <- 3
financialWB1$FWB2_3r[financialWB1$FWB2_3 == 4] <- 2
financialWB1$FWB2_3r[financialWB1$FWB2_3 == 5] <- 1

financialWB1$FWB2_4r <- NA
financialWB1$FWB2_4r[financialWB1$FWB2_4 == 1] <- 5
financialWB1$FWB2_4r[financialWB1$FWB2_4 == 2] <- 4
financialWB1$FWB2_4r[financialWB1$FWB2_4 == 3] <- 3
financialWB1$FWB2_4r[financialWB1$FWB2_4 == 4] <- 2
financialWB1$FWB2_4r[financialWB1$FWB2_4 == 5] <- 1
```

# Calculating and Interpreting Reliability

## *__goodFWB__ dataset*

### __Note:__ make sure that you don't have ggplot2 or tidyverse loaded in your kernel because that package has a different alpha() function that will mask the alpha() function in your psych package.

```r
alpha(goodFWB)
```

### *Inter-rater Reliability*
* The overall Chronbach's alpha is 0.88 (raw_alpha), which is an indicator of good reliability
* In the "Reliability if an item is dropped" section, all of the raw_alphas come under the original, so there is no advantage in dropping an item from the survey

### *Inter-item Reliability* 
* The correlation of all items with the rest of the scale (r.drop) > .3
* The "Non missing response frequency" table shows that the responses are relatively evenly distributed

##  *__badFWB__ dataset*

```r
alpha(badFWB)
```

### *Inter-rater Reliability*
* The overall Chronbach's alpha is 0.87 (raw_alpha), which suggests that this scale is reliable
* In the "Reliability if an item is dropped" section, all of the raw_alphas come under the original, so there is no advantage in dropping any one item from the survey

### *Inter-item Reliability* 
* The correlation of all items with the rest of the scale (r.drop) > .3
* The "Non missing response frequency" table shows that the responses are relatively evenly distributed

## *Interpreting the Financial Well Being Scale as a whole (__financialWB2__)*

### *Inter-rater Reliability*

### *Inter-item Reliability*

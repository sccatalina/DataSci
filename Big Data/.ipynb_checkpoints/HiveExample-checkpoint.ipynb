# HiveSQL Query

`create view AccidentsPerType as select vehicle_body_type, count(case_vehicle_id) as accidentCount
from mvc_10k
group by vehicle_body_type
order by accidentCount desc;`

`select vehicle_body_type, accidentCount
from AccidentsPerType;`
